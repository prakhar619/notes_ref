```python
import matplotlib.pyplot as plt
ar_freq = [0 for i in range(0,128)]
ar_Alpha = [chr(i) for i in range(0,128)]
```

```python
para = input()
para = para.upper()
for i in para:
        ar_freq[ord(i)]+=1
print(ar_freq)
```

```python
plt.bar([ar_Alpha[i] for i in range(65,91)],[ar_freq[i] for i in range(65,91)])
```
