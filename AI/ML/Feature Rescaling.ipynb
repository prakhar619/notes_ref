```python
import numpy as np

def mean_normalisation(x_data,ignore_feature_list):
# x_normalised_i = (x_i - x_mean) /(x_max - x_min)
    feature_mean = np.zeros(len(x_data[0]))
    feature_max = np.zeros(len(x_data[0]))
    feature_min = np.zeros(len(x_data[0]))
    feature_count = len(x_data[0])
    training_example_count = len(x_data)
    x_normalised = x_data
    
    flag = 1
    for i in range(training_example_count):
        for j in range(feature_count):
            if(not(j in ignore_feature_list)):
                feature_mean[j] += x_data[i][j]
                if(flag):
                    flag = 0
                    feature_max[j] = x_data[i][j]
                    feature_min[j] = x_data[i][j]
                else:
                    if(feature_max[j] < x_data[i][j]):
                        feature_max[j] = x_data[i][j]
                    if(feature_min[j] > x_data[i][j]):
                        feature_min[j] = x_data[i][j]
                    
    
    for i in range(feature_count):
        feature_mean[i]/=training_example_count
        
    for i in range(training_example_count):
        for j in  range(feature_count):
            if(not(j in ignore_feature_list)):
                x_normalised[i][j] = round((x_data[i][j] - feature_mean[j])/(feature_max[j] - feature_min[j]),3)
    
    return x_normalised
```

```python
#also called Standardization Scaling
def z_score_normalisation(x_data,ignore_feature_list):
# X_normalised_i = (x_i - x_mean) / x_stand_deviation
    feature_mean = np.zeros(len(x_data[0]))
    feature_deviation = np.zeros(len(x_data[0]))
    feature_count = len(x_data[0])
    training_example_count = len(x_data)
    x_normalised = x_data

    for i in range(training_example_count):
        for j in range(feature_count):
            if(not(j in ignore_feature_list)):
                feature_mean[j] += x_data[i][j]
    
    feature_mean/=training_example_count
    
    for i in range(training_example_count):
        for j in range(feature_count):
            if(not(j in ignore_feature_list)):
                feature_deviation[j] += (x_data[i][j] - feature_mean[j])**2
      
    feature_deviation/= training_example_count
    
    for i in range(len(feature_deviation)):
        feature_deviation[i] = feature_deviation[i]**(0.5)
    
    
    for i in range(training_example_count):
        for j in  range(feature_count):
            if(not(j in ignore_feature_list)):
                x_normalised[i][j] = round((x_data[i][j] - feature_mean[j])/(feature_deviation[j]),3)
    return x_normalised
    
```

```python
def csv_fileReader():
    #assumed no field empty
    #col indexing start from zero
    y_label_col = 8
    ignored_col = [9]
    filename = "./housing.csv"

    x_data = []
    y_label = []
    fields  = []
    import csv
    with open(filename,'r') as csvfile:
        csvreader = csv.reader(csvfile)
        flg = 1
        for row in csvreader:
            row_list = []
            feature_no = 0
            for item in row:
                if(feature_no == y_label_col):
                    if(flg == 0):
                        y_label.append(float(item))
                elif(not( feature_no in ignored_col)):
                    if(flg == 1):
                        fields.append(item)
                    else:
                        row_list.append(float(item))
                feature_no+=1
            if(flg == 1):
                flg = 0
            else:
                x_data.append(row_list)
    
    return x_data,y_label,fields

```
